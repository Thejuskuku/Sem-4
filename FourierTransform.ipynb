```python
import numpy as np
from skimage import io, color
import matplotlib.pyplot as plt

def fourier_transform(image_path):
    """
    Performs a Fourier Transform on an image, displays the magnitude spectrum,
    and reconstructs the image from the transform.

    Args:
        image_path (str): The path to the input image.
    """
    # Load the image in grayscale
    image = io.imread(image_path, as_gray=True)
    if image is None:
        print("Error: Could not read the image.")
        return

    # Compute the 2D Fourier Transform
    f_transform = np.fft.fft2(image)
    f_shift = np.fft.fftshift(f_transform)  # Shift the zero-frequency component to the center
    magnitude_spectrum = 20 * np.log(np.abs(f_shift) + 1)  # Add 1 to avoid log(0)

    # Compute the Inverse Fourier Transform
    f_ishift = np.fft.ifftshift(f_shift)
    image_reconstructed = np.fft.ifft2(f_ishift)
    image_reconstructed = np.abs(image_reconstructed)

    # Plot original image, magnitude spectrum, and reconstructed image
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 3, 1)
    plt.title("Original Image")
    plt.imshow(image, cmap='gray')
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.title("Magnitude Spectrum")
    plt.imshow(magnitude_spectrum, cmap='gray')
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.title("Reconstructed Image")
    plt.imshow(image_reconstructed, cmap='gray')
    plt.axis("off")

    plt.show()

# Example usage:
# Replace 'image.jpg' with the actual path to your image file
fourier_transform('image_path.jpg')
```
